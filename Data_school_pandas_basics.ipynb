```python
import pandas as pd 
#Importing first 
```

```python
orders = pd.read_table('http://bit.ly/chiporders')
#Reading a tabular data file from a URL.
```

```python
pd.read_table('http://bit.ly/movieusers')
```

```python
pd.read_table('http://bit.ly/movieusers',sep='|')
#Our separator here is | so we mention it within our read_table so that we get rid of it once downloading our tb.
```

```python
pd.read_table('http://bit.ly/movieusers',sep='|', header=None)
#Here the first row is being interpreted as the header row, and it's not. we have to tall pandas about it!
```

```python
user_cols = ['user_id', 'Age', 'Sex', 'job title', 'Postal code']
pd.read_table('http://bit.ly/movieusers',sep='|', header=None, names=user_cols)
#We create headers rows by creating a python list passing it as the value of the argument names.
#pandas know to use these strings as column names. 
```

```python
users = pd.read_table('http://bit.ly/movieusers',sep='|', header=None, names=user_cols)
#it is now a dataframe object 
```

```python
users.head()
#Reading the dataframe now looking clean and neat. 
```
