```python
import pandas as pd
```


‚òùüèª Reminder : A dataframe is a table of data and it has columns and rows. 
    


‚òùüèª Reminder : A series is a single column wihtin a dataframe. 
    

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.title.sort_values()
```

```python
type(movies.title.sort_values())
```


‚≠êÔ∏è By using 'sort_values()' method, We have indeed here sorted Series 'title' in ascending order. 



‚ûï It is also possible to sort a Series in descending order as follow: 


```python
movies.title.sort_values(ascending=False)
```


‚úÖ Our Series has then been successfully sorted in descending order, titles starting with a 'z' coming on top of the list.



‚ûï It is also possible to extract a Series in random order as follow : 
    

```python
movies['title']
```


‚úÖ Command successfuly performed. Extraction in initial order as shown into the reading of our dataframe.



‚≠êÔ∏è Here, we wish to sort the Series within the dataframe : 


```python
movies.sort_values('title')
```


‚úÖ This operation is successful as long you sort the dataframe by a specific column, and similarly it can be performed in descending order. 


```python
movies.sort_values('title', ascending=False)
```


‚úÖ This operation is successful as long you set the 'ascending' argument to false. Similarly, the sort can be done by selecting another Series.


```python
movies.sort_values('duration')
```


‚úÖ 'duration' Series is now sorted into ascending order. 


```python
movies.head()
```


üòâ Using the method 'head()', we verify that nothing has changed of the overall order.


 
 ‚≠êÔ∏è Let's see now how we could sort multiple columns. In our example bellow, we are going to sort two columns in one operation : 
 
 

```python
movies.sort_values(['content_rating', 'duration'])
```


‚úÖ it worked because we have passed a list of strings instead of a single string. As such, our two selected columns are sorted. It gets first sorted by content rating, and then by duration as per our request.
